Binary Diabetes Prediction

```python
#First import libary
import pandas as pd
```

```python
#Then import data
diatabetes = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Diabetes.csv')
```

```python
diatabetes.head()
```

```python
diatabetes.info()
```

```python
diatabetes.describe()
```

```python
#Define y and x
diatabetes.columns
```

```python
y = diatabetes['diabetes']
x = diatabetes[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

```python
#Splite ....
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=2529)
```

```python
#select  a model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=500) #by default is 100
```

```python
#Train    model
model.fit(x_train,y_train)
```

```python
#Now the time is Predict
f_pred = model.predict(x_test)
```

```python
# Last libary import for Acuracy
from sklearn.metrics import accuracy_score
```

```python
#Final Answer
accuracy_score(y_test,f_pred)
```
